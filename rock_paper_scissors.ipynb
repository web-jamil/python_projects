```python
import random

def print_rules():
    """Function to print the rules of the game."""
    print("Rock, Paper, Scissors Game")
    print("=================================")
    print("Choose your move:")
    print("1. Rock")
    print("2. Paper")
    print("3. Scissors")
    print("=================================")
    print("The rules are simple:")
    print("Rock beats Scissors.")
    print("Scissors beats Paper.")
    print("Paper beats Rock.")

def get_user_choice():
    """Function to get the user's choice."""
    while True:
        try:
            user_choice = int(input("Enter 1 for Rock, 2 for Paper, or 3 for Scissors: "))
            if user_choice in [1, 2, 3]:
                break
            else:
                print("Invalid choice, please enter 1, 2, or 3.")
        except ValueError:
            print("Invalid input, please enter a number.")
    return user_choice

def get_computer_choice():
    """Function to get the computer's random choice."""
    return random.choice([1, 2, 3])

def determine_winner(user_choice, computer_choice):
    """Function to determine the winner."""
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (user_choice == 1 and computer_choice == 3) or \
         (user_choice == 2 and computer_choice == 1) or \
         (user_choice == 3 and computer_choice == 2):
        return "You win!"
    else:
        return "Computer wins!"

def play_game():
    """Main game loop."""
    print_rules()

    while True:
        user_choice = get_user_choice()
        computer_choice = get_computer_choice()

        # Map the choices to their respective names
        choices = {1: "Rock", 2: "Paper", 3: "Scissors"}
        
        print(f"\nYou chose: {choices[user_choice]}")
        print(f"Computer chose: {choices[computer_choice]}")
        
        result = determine_winner(user_choice, computer_choice)
        print(result)

        play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()
        if play_again != "yes":
            print("Thanks for playing! Goodbye!")
            break

if __name__ == "__main__":
    play_game()
```

```python
import random

class RockPaperScissorsGame:
    def __init__(self):
        self.choices = {1: "Rock", 2: "Paper", 3: "Scissors"}
        self.rules = {
            (1, 2): "Paper beats Rock.",
            (2, 3): "Scissors beats Paper.",
            (3, 1): "Rock beats Scissors."
        }

    def print_rules(self):
        """Prints the game rules."""
        print("Rock, Paper, Scissors Game")
        print("=================================")
        print("Choose your move:")
        print("1. Rock")
        print("2. Paper")
        print("3. Scissors")
        print("=================================")
        print("The rules are simple:")
        print("Rock beats Scissors.")
        print("Scissors beats Paper.")
        print("Paper beats Rock.")

    def get_user_choice(self):
        """Gets the user's choice, ensures it's valid."""
        while True:
            try:
                user_choice = int(input("Enter 1 for Rock, 2 for Paper, or 3 for Scissors: "))
                if user_choice in self.choices:
                    break
                else:
                    print("Invalid choice, please enter 1, 2, or 3.")
            except ValueError:
                print("Invalid input, please enter a number.")
        return user_choice

    def get_computer_choice(self):
        """Generates a random choice for the computer."""
        return random.choice([1, 2, 3])

    def determine_winner(self, user_choice, computer_choice):
        """Determines the winner of the round."""
        if user_choice == computer_choice:
            return "It's a tie!"
        elif (user_choice, computer_choice) in self.rules:
            return f"You win! {self.rules[(user_choice, computer_choice)]}"
        else:
            return f"Computer wins! {self.rules[(computer_choice, user_choice)]}"

    def play_round(self):
        """Plays a single round of the game."""
        user_choice = self.get_user_choice()
        computer_choice = self.get_computer_choice()

        print(f"\nYou chose: {self.choices[user_choice]}")
        print(f"Computer chose: {self.choices[computer_choice]}")

        result = self.determine_winner(user_choice, computer_choice)
        print(result)

    def play_game(self):
        """Main game loop."""
        self.print_rules()

        while True:
            self.play_round()
            play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()
            if play_again != "yes":
                print("Thanks for playing! Goodbye!")
                break


if __name__ == "__main__":
    game = RockPaperScissorsGame()  # Create an instance of the game
    game.play_game()  # Start the game
```
